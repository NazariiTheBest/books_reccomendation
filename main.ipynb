```python
import zipfile
import json
import pandas as pd

books = []

with zipfile.ZipFile("C:\\Users\\ohrim\\PycharmProjects\\tg_parse\\kpi_nlp_course\\lab2\\Data\\yakaboo-book-reviews-dataset\\books.jsonlines.zip") as zip_ref:
    with zip_ref.open(zip_ref.namelist()[0]) as f:
        for line in f:
            decoded_line = line.decode("utf-8", errors="replace")
            try:
                books.append(json.loads(decoded_line))
            except json.JSONDecodeError as e:
                print("Помилка JSON:", e)
                continue

df = pd.DataFrame(books)
```

```python
reviews = []

with zipfile.ZipFile("C:\\Users\\ohrim\\PycharmProjects\\tg_parse\\kpi_nlp_course\\lab2\\Data\\yakaboo-book-reviews-dataset\\reviews.jsonlines.zip") as zip_ref:
    with zip_ref.open(zip_ref.namelist()[0]) as f:
        for line in f:
            decoded_line = line.decode("utf-8", errors="replace")
            try:
                reviews.append(json.loads(decoded_line))
            except json.JSONDecodeError as e:
                print("Помилка JSON:", e)
                continue

df_reviews = pd.DataFrame(reviews)
```

```python
df["features"] = df["categories"].apply(lambda x: " ".join(x))
# df["features"] = df["title"] + " " + df["description"] + " " + df["features"]
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
word_matrix = vectorizer.fit_transform(df['features'])
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def get_recommendation(title: str, df: pd.DataFrame,):
    
    index = df.index[df["title"].str.lower() == title.lower()]
    
    query_vector = vectorizer.transform([title])
    word_matrix = vectorizer.transform(df["features"])
    print(query_vector)
    
    similarities = cosine_similarity(query_vector, word_matrix)[:, 1:]
    print(similarities.shape)
    
    return df.iloc[np.argmax(similarities)]
    
```

```python
df_merge = df.merge(df_reviews, on="book_id")
df_merge.head()
```

```python
df_merge["features"] = df_merge["categories"].apply(lambda x: " ".join(x)) + df_merge["review"] + df_merge["reviewTitle"]
word_matrix = vectorizer.fit_transform(df_merge["features"])
```

```python
query = vectorizer.transform(["У пошуках Івана Сили"])
sim = cosine_similarity(query, word_matrix)[:, 1:]

np.argmax(sim)
df_merge.iloc[np.argmax(sim)]
```

```python
df_merge = df_merge.groupby("title")[["titleEnglish", "description", "attrs", "rating_x", "review", "book_id", "categories"]].agg(
    {
    "titleEnglish": "first",
    "description": "first",
    "attrs": "first",
    "rating_x": "first",
    "review": lambda x: " ".join(x),
    "book_id": "first",
    "categories": "first",
    }
)
df_merge = df_merge.rename({"rating_x": "rating"}, axis=1)
df_merge = df_merge.sort_values("book_id", ascending=True)
df_merge.head()
```

```python
df_merge["features"] = df_merge["categories"].apply(lambda x: " ".join(x)) + df_merge["review"]
```

```python
def get_recommendation(query_str: str, df: pd.DataFrame, vectorzer: TfidfVectorizer):
    
    word_matrix = vectorizer.fit_transform(df["features"])
    query = vectorizer.transform([query_str])
    similarities = cosine_similarity(query, word_matrix).flatten()
    similarities[0] = -1
    top5_indices = np.argpartition(-similarities, 5)[:5]
    top5_indices_sorted = top5_indices[np.argsort(-similarities[top5_indices])]
    print(top5_indices_sorted)
    print(similarities[top5_indices_sorted])
    
    top5_rows = df.iloc[top5_indices_sorted]
    return top5_rows


get_recommendation("У пошуках Івана Сили", df_merge, vectorizer)
```
